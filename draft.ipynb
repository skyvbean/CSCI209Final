For the final project, our group will create a turn-based combat system (inspired by games like Pokemon). It will use and teach boolean algebra topics primarily.

**Boolean algebra will be used to detect game states and handle the enemy’s behavior.** It will also be employed to **define moves or attacks**. In other words, to help tie this project to boolean algebra and teach its concepts, actions or attacks will follow logical statements. For example, the project will feature a **special attack that will successfully execute if and only if both the player and enemy select an attack for that turn.**

Examples:

Enemy Behavior  
* The enemy will use game variables to determine what choice to make. If we include multiple behavior types, one enemy might mainly select its most reliable action while another picks its strongest action. If the player is low on health and its strongest move deals more damage than the player’s health value, it will select that move. Similarly, if the enemy has low health, it may choose to heal if it has a healing action in its moveset.  

Player Moves  
* For this project, many of the player’s stronger moves will follow logical rules to successfully perform. Like Pokemon this project may include a simple **type bonus system where certain types of moves are more or less effective to specific enemies for more boolean algebra application**.
