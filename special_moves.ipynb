```python
# COUNTER-ATTACK

def counter_attack(actor, data):
    # 1.3x damage IF enemy attacks this turn; MISS otherwise.
    
    # get attacker and defender
    if actor == "player":
        defender_action = data["enemy_action"]
        base_dmg = data["player_type"]["attack_base_hp"]
        dmg_target = "enemy_hp"
    else:
        defender_action = data["player_action"]
        base_dmg = data["enemy_type"]["attack_base_hp"]
        dmg_target = "player_hp"
    # move logic
    if defender_action == "ATTACK":
        damage = int(base_dmg * 1.3)
        data[dmg_target] -= damage
        return f"{actor.upper()} Counter-Attacks for {damage} damage!"
    else:
        return f"{actor.upper()}'s Counter-Attack MISSED (enemy did not attack)!"
```

```python
# LIFESTEAL

def lifesteal(actor, data):

    # IF enemy has >60% HP AND you have <30% HP â†’ heal 1/2 base HP.


    if actor == "player":
        me_hp = data["player_hp"]
        me_type = data["player_type"]
        enemy_hp = data["enemy_hp"]
        enemy_type = data["enemy_type"]
    else:
        me_hp = data["enemy_hp"]
        me_type = data["enemy_type"]
        enemy_hp = data["player_hp"]
        enemy_type = data["player_type"]

    # Compute thresholds
    enemy_60 = enemy_type["base_hp"] * 0.60
    me_30 = me_type["base_hp"] * 0.30

    if enemy_hp > enemy_60 and me_hp < me_30:
        heal = me_type["heal_base_hp"] // 2
        if actor == "player":
            data["player_hp"] += heal
        else:
            data["enemy_hp"] += heal
        return f"{actor.upper()} Lifesteals {heal} HP!"
    else:
        return f"{actor.upper()}'s Lifesteal had no effect."
```
